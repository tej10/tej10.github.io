```python
# frequency distribution of words in a text
text='"ఆడిట్ నిర్వహణ ఆడిటర్ ఒక కొత్త ఆడిట్ చేపట్టే ముందు సక్రమ పద్ధతి లో కార్య ప్రణాళికను రూపొందించాలి . దాని కనుగుణంగా వ్యవహరించాలి . పత్రసహిత సాక్ష్యాధారాల తో , వ్యవహారాల ను తనిఖీ చేయాలి . ఆడిట్ చేసే విధానం సంస్థ అవసరాల ను బట్టి , అంతర్గత తనిఖీన్ బట్టి , ఇంకా అనేక ఇతర విషయాల ను బట్టి మారుతూఉంటుంది . పద్దుపుస్తకాలలో ని అంకగణిత యదార్థత ను , సరిచూడటాని కి , సాక్ష్యాలు సమాచారం , వివరణలు , అవసరమౌతాయి . సాక్ష్యాలు వోచర్లరూపం లో , లేదా పత్రాల రూపం లో ఉండవచ్చు . కావలసిన సమాచారాన్ ని వివరణల ను ఆడిటర్ "'
fd=nltk.FreqDist(text.split())
fd
```

```python
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in text.split())
#list of conditons
cfd.conditions()
```

```python
cfd[4]
```

```python
indiandata = indian.words(fileids = 'telugu.pos')
cdf = ConditionalFreqDist((len(word), word) for word in indiandata)
cdf[4]
```

```python
data = []
for words in indiandata:
    if len(words)>5:
        data.append(words)
```

```python
fd = nltk.FreqDist(data)
fd
```

```python
sorted_fd = sorted(fd.items(), key = lambda x:x[1])
```

```python
print(sorted_fd)
```

```python
word = list(sorted_fd)[len(sorted_fd)-1]
```

```python
print("most frequently used word in this document is:", word)
```
