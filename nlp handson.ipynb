```python
fd
```

```python
#Task 6: Conditional Frequency Distribution of words in a text
#tells us how many 2 letter words or 3 letter words 
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word), word) for word in text1.split())
cfd[3]
```

```python
#Task 2 - SIMPLE TEXT CLASSIFIER
def gender_features(word):
    return{'last_letter':word[-1]}
gender_features('Charles')
```

```python
from nltk.corpus import names
labeled_names = ([(name, 'male') for name in names.words('male.txt')]+[(name, 'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
```

```python
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names]
```

```python
train_set, test_set = featuresets[500:],featuresets[:500]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Tyrion'))
```

```python
classifier.classify(gender_features('Jaimiefoxx'))
```

```python
classifier.classify(gender_features('Monica'))
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
# Tool names for positive and negative
#TEXTBLOB
#Jensen
# CMU Sphinx
#Task3: Vectorizers & Cosine Similarity
# CountVectorizer - Learn
# TfidfVectorizer
# from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# fit method vectorizes all the unique words in the corpus
vect = CountVectorizer(binary = True)
corpus = ["Paul Edgecomb, the head guard of a prison, meets an inmate, John Coffey, a black man who is accused of murdering two girls.", "His life changes drastically when he discovers that John has a special gift."]
vect.fit(corpus)
```

```python
vocab = vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
    print("{}:{}".format(key,vocab[key]))
```

```python
# do plagarism task using vectorization
print(vect.transform(["Be yourself, everyone else is taken"]).toarray())
```

```python
print(vect.transform(corpus).toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vect.transform(["Be yourself, that is it"]).toarray(), vect.transform(["I don't know what to type"]).toarray())
```

```python
print(similarity[0][0])
```
